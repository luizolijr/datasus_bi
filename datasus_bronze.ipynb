# CONCATENAR ARQUIVOS CSV

```python
dataframes_2012 = []
dataframes_2022 = []

datas_2012 = ['2012-01-01', '2012-02-01', '2012-03-01', '2012-04-01', '2012-05-01', '2012-06-01', '2012-07-01', '2012-08-01', '2012-09-01', '2012-10-01', '2012-11-01', '2012-12-01', '2013-01-01', '2013-02-01', '2013-03-01', '2013-04-01', '2013-05-01', '2013-06-01', '2013-07-01', '2013-08-01', '2013-09-01']
datas_2022 = ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01', '2022-06-01', '2022-07-01', '2022-08-01', '2022-09-01', '2022-10-01', '2022-11-01', '2022-12-01', '2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01', '2023-09-01']

datas_2012 = [i[2:4]+i[5:7]+'.csv' for i in datas_2012]
datas_2022 = [i[2:4]+i[5:7]+'.csv' for i in datas_2022]

for i in datas_2012:
    filename = path+'RDRJ'+i
    dados = pd.read_csv(filename, delimiter=';', encoding='latin-1')
    dataframes_2012.append(dados)

dados_2012 = pd.concat(dataframes_2012)
dados_2012.to_csv('201201_201309.csv', index=False)


for i in datas_2022:
    filename = path+'RDRJ'+i
    dados = pd.read_csv(filename, delimiter=';', encoding='latin-1')
    dataframes_2022.append(dados)

dados_2022 = pd.concat(dataframes_2022)
dados_2022.to_csv('202201_202309.csv', index=False)
```

# Função de Tratamento

```python
def tratamento(dados):
    
    dados['diag_princ'] = [i for i in dados['DIAG_PRINC']]
    dados['DIAG_PRINC'] = [i[:3] for i in dados['DIAG_PRINC']]
    
    dados = pd.merge(dados, cid10_detailed, left_on='diag_princ', right_on='SUBCAT', how='left')
    dados = pd.merge(dados, cid10, left_on='DIAG_PRINC', right_on='codigo', how='left')

    dados['SEXO'] = dados['SEXO'].apply(lambda x: 'Masculino' if x == 1 else ('Feminino' if x == 2 else ('Feminino' if x == 3 else 'erro')))

    return dados
    
# df_final = merge_detailed[['SEXO','IDADE', 'diag_princ', 'DESCRICAO', 'VAL_TOT', 'VAL_SH', 'VAL_SP', 'MORTE']]
# test_detailed = test_detailed[test_detailed['diag_princ'].str[0] == 'B']
# test_detailed = test_detailed[test_detailed['diag_princ'].str[1] == '1']
# test_detailed = test_detailed[test_detailed['diag_princ'].str[2] == '6']

# test_detailed
```

# Ler arquivo concatenado (todos os dados)


```python
path_arquivo_completo = 'C:\\Users\\luizjr\\REPOS\\data_sus\\'

df_2012 = pd.read_csv(path_arquivo_completo+'201201_201309.csv', delimiter=',', encoding='latin1')
df_2022 = pd.read_csv(path_arquivo_completo+'202201_202309.csv', delimiter=',', encoding='latin1')

df_2012
```

# Rodar função e salvar o csv


```python
df_2012 = tratamento(df_2012)
df_2022 = tratamento(df_2022)


df_2012.to_csv('201201_201309_tratado.csv', index=False)
df_2022.to_csv('202201_202309_tratado.csv', index=False)
```
